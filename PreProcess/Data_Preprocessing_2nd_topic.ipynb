```python
# 첫 번째 열에서 특정 키워드를 포함하는 행을 필터링하여 출력하는 함수 정의
def print_rows_with_keyword(df, keyword):
    filtered_rows = df[df.iloc[:, 0].str.contains(keyword, na=False)]
    print(filtered_rows)

print("welfare.xlsx:")
print_rows_with_keyword(welfare, '제')
```

```python
def remove_keyword_from_specific_cells(df, specific_contents, keyword_to_remove):
    # 여러 특정 내용을 포함하는 행 필터링
    mask = df.iloc[:, 0].apply(lambda x: any(content in x for content in specific_contents) if pd.notna(x) else False)
    # 조건에 맞는 셀에서만 키워드 삭제
    df.loc[mask, df.columns[0]] = df.loc[mask, df.columns[0]].str.replace(keyword_to_remove, '', regex=False)
    return df

# 원하는 내용 리스트를 포함하는 셀에서 '제'라는 단어를 삭제
specific_contents = ['길음제2동', '면목제3.8동', '상일제1동', '상일제2동', '자양제4동']
keyword_to_remove = '제'

welfare = remove_keyword_from_specific_cells(welfare, specific_contents, keyword_to_remove)

# 함수 정의: 첫 번째 열에서 '-'와 ','를 '.'로 변경
def replace_characters(df, old_chars, new_char):
    for old_char in old_chars:
        df.iloc[:, 0] = df.iloc[:, 0].str.replace(old_char, new_char, regex=False)
    return df

# 바꾸고자 하는 문자 리스트와 새로운 문자
old_chars = ['-', ',']
new_char = '.'

welfare = replace_characters(welfare, old_chars, new_char)

welfare.to_excel('기초생활수급자_동별_전처리_2nd.xlsx', index=False)
```

```python
# 변경하고자 하는 특정 내용 목록
contents_to_replace = ['정능1동', '정능2동', '정능3동', '정능4동']
old_word = '능'
new_word = '릉'

# 특정 내용에서 단어를 변경하는 함수 정의
def replace_specific_words(df, contents, old_word, new_word):
    for content in contents:
        mask = df.iloc[:, 0].str.contains(content, na=False)
        df.loc[mask, df.columns[0]] = df.loc[mask, df.columns[0]].str.replace(old_word, new_word, regex=False)
    return df

# 데이터프레임에서 단어를 변경
alone = replace_specific_words(alone, contents_to_replace, old_word, new_word)

alone.to_excel('독거노인(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
```

```python
# 삭제하고자 하는 특정 내용 리스트
contents_to_remove = ['일원2동', '개포3동']  # 삭제하고 싶은 내용을 여기에 추가하세요

# 첫 번째 열에서 특정 내용을 포함하는 행을 삭제하는 함수 정의
def remove_rows_with_specific_content(df, contents):
    mask = df.iloc[:, 0].apply(lambda x: any(content in x for content in contents) if pd.notna(x) else False)
    return df[~mask].reset_index(drop=True)

# 각 데이터프레임에서 특정 내용을 포함하는 행 삭제
welfare = remove_rows_with_specific_content(welfare, contents_to_remove)
alone = remove_rows_with_specific_content(alone, contents_to_remove)
pop = remove_rows_with_specific_content(pop, contents_to_remove)
disabled = remove_rows_with_specific_content(disabled, contents_to_remove)

# 각 전처리된 데이터를 저장
welfare.to_excel('기초생활수급자_동별_전처리_2nd.xlsx', index=False)
alone.to_excel('독거노인(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
pop.to_excel('주민등록인구(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
disabled.to_excel('장애인(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
```

```python
# 각 데이터프레임에서 첫번째 열의 내용중 '신사동' 행을 합치기
disabled_rows = disabled[disabled.iloc[:, 0].str.contains('신사동', na=False)]
disabled_merged = pd.DataFrame(disabled_rows.sum(axis=0)).T

alone_rows = alone[alone.iloc[:, 0].str.contains('신사동', na=False)]
alone_merged = pd.DataFrame(alone_rows.sum(axis=0)).T

pop_rows = pop[pop.iloc[:, 0].str.contains('신사동', na=False)]
pop_merged = pd.DataFrame(pop_rows.sum(axis=0)).T

# 합쳐진 행을 원래 데이터프레임에서 삭제
disabled = disabled[~disabled.iloc[:, 0].str.contains('신사동', na=False)]
alone = alone[~alone.iloc[:, 0].str.contains('신사동', na=False)]
pop = pop[~pop.iloc[:, 0].str.contains('신사동', na=False)]

# 데이터프레임에 새로운 행 추가
disabled = pd.concat([disabled, disabled_merged], ignore_index=True)
pop = pd.concat([pop, pop_merged], ignore_index=True)
alone = pd.concat([alone, alone_merged], ignore_index=True)

# 합쳐진 이름 수정
disabled.iloc[:, 0] = disabled.iloc[:, 0].str.replace('신사동신사동', '신사동')
pop.iloc[:, 0] = pop.iloc[:, 0].str.replace('신사동신사동', '신사동')
alone.iloc[:, 0] = alone.iloc[:, 0].str.replace('신사동신사동', '신사동')

# 전처리된 데이터 저장
pop.to_excel('주민등록인구(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
disabled.to_excel('장애인(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
alone.to_excel('독거노인(연령별_동별)_2022_전처리_2nd.xlsx', index=False)
```

```python
import pandas as pd
import re

# pop 데이터프레임의 첫 번째 열의 이름 가져오기
first_column_name = pop.columns[0]

# 각 데이터프레임의 첫 번째 열에서 '-', ',', '.' 문자 제거
pop[first_column_name] = pop[first_column_name].astype(str).apply(lambda x: re.sub(r'[-,.\d]', '', x))
alone[first_column_name] = alone[first_column_name].astype(str).apply(lambda x: re.sub(r'[-,.\d]', '', x))
disabled[first_column_name] = disabled[first_column_name].astype(str).apply(lambda x: re.sub(r'[-,.\d]', '', x))
welfare[first_column_name] = welfare[first_column_name].astype(str).apply(lambda x: re.sub(r'[-,.\d]', '', x))

# 각 데이터프레임을 첫 번째 열을 기준으로 그룹화하고 다른 열들을 합산하여 재설정
pop = pop.groupby(first_column_name).sum().reset_index()
alone = alone.groupby(first_column_name).sum().reset_index()
disabled = disabled.groupby(first_column_name).sum().reset_index()
welfare = welfare.groupby(first_column_name).sum().reset_index()

# 각 데이터프레임을 엑셀 파일로 저장
welfare.to_excel('기초생활수급자_동별_전처리_3rd.xlsx', index=False)
alone.to_excel('독거노인(연령별_동별)_2022_전처리_3rd.xlsx', index=False)
pop.to_excel('주민등록인구(연령별_동별)_2022_전처리_3rd.xlsx', index=False)
disabled.to_excel('장애인(연령별_동별)_2022_전처리_3rd.xlsx', index=False)
```
