```python
import pandas as pd
import re

# 엑셀 파일 읽기
pop = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/주민등록인구(연령별_동별)_2022_전처리.xlsx')

# 첫 번째 행을 헤더로 설정
header = pop.iloc[0]
pop_rest = pop.iloc[1:]

# 첫 번째 열을 기준으로 정렬
pop_sorted = pop_rest.sort_values(by=pop_rest.columns[0])

# 헤더와 나머지 데이터를 결합
pop = pd.concat([header.to_frame().T, pop_sorted], ignore_index=True)

# 첫 번째 행 저장
first_row = pop.iloc[0]

# 숫자와 마침표를 제거하는 함수 정의
def remove_numbers_and_periods(text):
    return re.sub(r'\d+', '', text)

# 첫 번째 열에서 숫자와 마침표 제거
pop.iloc[:, 0] = pop.iloc[:, 0].apply(remove_numbers_and_periods)

# 나머지 열들을 숫자형으로 변환
pop.iloc[:, 1:] = pop.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# 행을 병합하는 함수 정의
def merge_rows(group):
    first_value = group.iloc[0, 0]
    sum_values = group.iloc[:, 1:].sum()
    return pd.Series([first_value] + sum_values.tolist(), index=['First Column'] + pop.columns[1:].tolist())

# 첫 번째 열을 기준으로 그룹화하여 행 병합
pop = pop.groupby(pop.columns[0]).apply(merge_rows).reset_index()

# 첫 번째 행을 데이터프레임에 다시 추가
pop = pd.concat([first_row.to_frame().T, pop])

# 네 번째 열 삭제
pop.drop(pop.columns[[4]], axis=1, inplace=True)

# 결과를 엑셀 파일로 저장
pop.to_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/주민등록인구(연령별_동별)_2022_전처리_fin.xlsx', index=False)
```

```python
import pandas as pd
import re

# CSV 파일 경로 설정
csv_file_path = "/content/drive/MyDrive/Data_mining/team_project/data_2022/서울시 경로당 정보.csv"

# CSV 파일 읽기, 인코딩 방식 처리
try:
    center = pd.read_csv(csv_file_path, encoding='euc-kr')
except UnicodeDecodeError:
    center = pd.read_csv(csv_file_path, encoding='cp949')

# 첫 두 열 삭제
center.drop(center.columns[0:2], axis=1, inplace=True)

# 두 번째 열부터 마지막 열까지 삭제
center.drop(center.columns[1:], axis=1, inplace=True)

# 열 이름 변경: '소재지전체주소'를 'CT_NM'으로 변경
center.rename(columns={'소재지전체주소': 'CT_NM'}, inplace=True)

# 주소에서 구 이름 제거하는 함수 정의
def remove_district_name(address):
    districts = ['동작구', '성동구', '강동구', '동대문구']
    for district in districts:
        address = address.replace(district, '')
    return address.strip()

# 구 이름 제거
center['CT_NM'] = center['CT_NM'].apply(remove_district_name)

# 주소에서 '동' 또는 '가'를 추출하는 함수 정의
def extract_district_or_ga(address):
    match = re.search(r'\b\w*[동가]\w*\b', address)
    return match.group(0) if match else ''

# '동' 또는 '가' 추출
center['CT_NM'] = center['CT_NM'].apply(extract_district_or_ga)

# 'CT_NM' 기준으로 그룹화하여 첫 번째 값으로 병합하고 중복 제거
center = center.groupby('CT_NM').first().reset_index()
center.drop_duplicates(inplace=True)

# 첫 번째 행 삭제
center.drop(center.index[0], inplace=True)

print(center.head())
print(center.info())

# 전처리된 데이터를 새로운 엑셀 파일로 저장
output_file = "/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/경로당_동별_전처리.xlsx"
center.to_excel(output_file, index=False)
```

```python
import pandas as pd
import re

# 전처리된 엑셀 파일 읽기
center = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/경로당_동별_전처리.xlsx')

# 숫자를 제거하는 함수 정의
def remove_numbers(input_string):
    return re.sub(r'\d', '', str(input_string))

# 첫 번째 열에서 숫자 제거
center.iloc[:, 0] = center.iloc[:, 0].apply(lambda x: remove_numbers(x) if isinstance(x, str) else x)

# 'CT_NM' 열의 값이 '가'로 끝나면 마지막 글자 제거
for i, value in enumerate(center['CT_NM']):
    if str(value).endswith('가'):
        center.at[i, 'CT_NM'] = value[:-1]

# 'CT_NM' 열을 기준으로 그룹화하고 중복 제거
center = center.groupby('CT_NM').agg(lambda x: ' '.join(set(x))).reset_index()

# 전처리된 데이터를 새로운 엑셀 파일로 저장
center.to_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/경로당_동별_전처리_fin.xlsx', index=False)
```

```python
import pandas as pd

# CSV 파일 경로 설정
csv_file_path = "/content/drive/MyDrive/Data_mining/team_project/data_2022/서울특별시 국민기초생활 수급자 동별 현황_20210731_v1.csv"

# CSV 파일 읽기, 인코딩 방식 처리
try:
    welfare = pd.read_csv(csv_file_path, encoding='euc-kr')
except UnicodeDecodeError:
    welfare = pd.read_csv(csv_file_path, encoding='cp949')

# '18세미만' 및 '18~64세' 데이터를 제외
welfare = welfare[~welfare.iloc[:, 4].isin(['18세미만', '18~64세'])]

# 첫 두 열 삭제
welfare.drop(welfare.columns[0:2], axis=1, inplace=True)

# '구'로 끝나는 단어 확인 (주석 처리된 부분은 필요 시 사용)
# filtered_words = welfare[welfare.iloc[:, 0].str.endswith('구')].iloc[:, 0].unique()
# for word in filtered_words:
#     print(word)

# '구'로 끝나는 행 제외
welfare = welfare[~welfare.iloc[:, 0].str.endswith('구')]

# 두 번째 열과 세 번째 열 삭제
welfare.drop(welfare.columns[1:3], axis=1, inplace=True)

# 첫 번째 열을 기준으로 그룹화하여 합계 계산
welfare = welfare.groupby(welfare.columns[0], as_index=False).sum()

# 열 이름 변경
welfare.rename(columns={'읍면동': 'CT_NM', '수급권자수': 'over_65'}, inplace=True)

print(welfare.head())
print(welfare.info())

# 전처리된 데이터를 새로운 엑셀 파일로 저장
output_file = "/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/기초생활수급자_동별_전처리.xlsx"
welfare.to_excel(output_file, index=False)
```

```python
import pandas as pd
import re

# 전처리된 엑셀 파일 읽기
welfare = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/기초생활수급자_동별_전처리.xlsx')

# 헤더와 데이터 분리
header = welfare.iloc[0]
welfare_rest = welfare.iloc[1:]

# 첫 번째 열을 기준으로 데이터 정렬
welfare_sorted = welfare_rest.sort_values(by=welfare_rest.columns[0])

# 헤더와 정렬된 데이터를 다시 합침
welfare = pd.concat([header.to_frame().T, welfare_sorted], ignore_index=True)

# 첫 번째 행 저장
first_row = welfare.iloc[0]

# 숫자와 마침표를 제거하는 함수 정의
def remove_numbers_and_periods(text):
    return re.sub(r'\d+', '', text)

# 첫 번째 열에서 숫자와 마침표 제거
welfare.iloc[:, 0] = welfare.iloc[:, 0].apply(remove_numbers_and_periods)

# 'CT_NM' 열에서 '-'와 '.' 제거
welfare['CT_NM'] = welfare['CT_NM'].str.replace('-', '')
welfare['CT_NM'] = welfare['CT_NM'].str.replace('.', '')

# 나머지 열을 숫자로 변환
welfare.iloc[:, 1:] = welfare.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

# 행을 병합하는 함수 정의
def merge_rows(group):
    first_value = group.iloc[0, 0]
    sum_values = group.iloc[:, 1:].sum()
    return pd.Series([first_value] + sum_values.tolist(), index=['First Column'] + welfare.columns[1:].tolist())

# 첫 번째 열을 기준으로 그룹화하고 행 병합
welfare = welfare.groupby(welfare.columns[0]).apply(merge_rows).reset_index()

# 첫 번째 행을 다시 데이터프레임에 추가
welfare = pd.concat([first_row.to_frame().T, welfare])

# 필요 없는 열 삭제
welfare.drop(welfare.columns[[2]], axis=1, inplace=True)

# 'CT_NM' 열에서 '제' 단어 제거
welfare['CT_NM'] = welfare['CT_NM'].str.replace(r'\b제\b', '', regex=True)

# 전처리된 데이터를 새로운 엑셀 파일로 저장
welfare.to_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/기초생활수급자_동별_전처리_fin.xlsx', index=False)
```

```python
import pandas as pd
import re

pop = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/pop.xlsx')
alone = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/alone.xlsx')
disabled = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/disabled.xlsx')
welfare = pd.read_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/welfare.xlsx')
```

```python
#pop 데이터의 첫번째 열의 내용 가져오기
first_column = pop.iloc[:, 0]

#pop 데이터의 첫번째 열의 내용을 'CT_NM'으로 생성
new_data = pd.DataFrame({'CT_NM': first_column})

# feature(독거노인 비율) 생성
alone_ratio = alone['all'] / pop['65_over']
new_data['alone_ratio'] = alone_ratio

# feature(노인 비율) 생성
old_ratio = pop['65_over'] / pop['all']
new_data['old_ration'] = old_ratio

# feature(노인 중 기초생활수급자 비율) 생성
welfare_ratio = welfare['over_65'] / pop['65_over']
new_data['welfare_ratio'] = welfare_ratio

# feature(노인중 장애인 비율) 생성
disabled_ratio = disabled['60_over'] / pop['60_over']
new_data['disabled_ratio'] = disabled_ratio

# 생성된 데이터파일 저장
new_data.to_excel('/content/drive/MyDrive/Data_mining/team_project/data_2022/using_data/final_data/new_data.xlsx', index=False)
```
